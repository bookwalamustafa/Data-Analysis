## Exercise 2:
Plot a histogram of the median_income column. What are the mean and median values of median_income? Write a brief discussion of your insights about the results in 100 words or less.

````python
df['median_income'].plot.hist()  # Plotting the histogram for the 'median_income' column
print(f"Mean value of the Median Income Column: {df['median_income'].mean()}")  # Printing the Mean value of the 'median_income' column
print(f"Median value of Median Income column: {df['median_income'].median()}")  # Printing the Median value of the 'median_income' column

# The one thing that I can infer from this is that the mean and median values are pretty close to each other which is a good sign showcasing that there is not a high probability for outliers in the dataset. Additionally, the frequency for the higher median_incomes gets significantly less is also a key observation.
````

## Exercise 3:

Plot the distribution of median_house_value in box plot. What are the median, 25% percentile, 75% percentile, and mean of the median_house_value? Write a brief discussion of your insights about the results in 100 words or less.

````python
df['median_house_value'].plot.box()  # Plotting the box plot of the 'median_house_value' column
plt.ylabel("House Value")  # Setting label for y-axis
print(f"Mean value of the Median House Value Column: {df['median_house_value'].mean()}")  # Printing the Mean value of the 'median_house_value' column
print(f"Median value of the Median House Value Column: {df['median_house_value'].median()}")  # Printing the Median value of the 'median_house_value' column
print(f"25% percentile value of the Median House Value Column: {df['median_house_value'].quantile(0.25)}")  # Printing the 25% percentile value of the 'median_houes_value' column
print(f"75% percentile value of the Median House Value Column: {df['median_house_value'].quantile(0.75)}")  # Printing the 75% percentile value of the 'median_house_value' column

# I see that most of the data is somewhere under its 75% percentile mark. Even in the median value, the data is more inclined towards the 25% percentile mark than the 75% mark.
# What I also infer is that there are some median_house_value somewhere around 500000 while most of the median_house_value is near the values mentioend below.
````

## Exercise 4:

Plot the distribution of ocean_proximity into bar plot. Write a brief discussion of your insights about the results in 100 words or less.

````python
df['ocean_proximity'].value_counts().plot.bar()  # Plotting bar plots for the number of values for all the entries in the 'ocean_proximity' column using the value_counts function
plt.ylabel("Number of houses per ocean proximity")  # Setting label for y-axis

# From the below bar plot, the one thing that I infer is that the '<1H OCEAN value' has the highest number in that column with 'ISLAND' has almost close to 0 values. That means that most of the houses that the data is showing is only less than one hour away from the ocean and very less houses on an island.
````
