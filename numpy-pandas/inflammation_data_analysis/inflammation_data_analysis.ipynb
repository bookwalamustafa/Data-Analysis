## Question 2 [20 marks]:
Download the zipped file "inflammation-data.zip" and unzip it in a folder. You should see a directory "inflammation-data" containing data files. Load the data in the file 'inflammation-05.csv' to an Numpy array.

````python
file_path = '/content/drive/MyDrive/Colab Notebooks/INFO 212/Week 3/Assignment 1/inflammation-data/inflammation-05.csv'  # Loading the path of the file

arr = np.loadtxt(file_path, delimiter=',')  # Loading the file using a delimeter into a Numpy array
print(arr)  # Displaying its contents
````

1. Suppose the rows represent patients, and columns represent days. How many patients and how many days does the file contain?

````python
print(f"The data represents {(arr.shape)[0]} patients.")  # Displaying the number of patients/rows
print(f"The data represents {(arr.shape)[1]} days.")  # Displaying the number of days/columns
````

2. Extract and display the values of the last 10 patients' inflammation in last 10 days.

````python
last_10_patients = arr[50:, 30:]  # Storing only the last 10 patients/rows data for the last 10 days/columns

for i in range(0, len(last_10_patients+1)):  # Running a loop for every single of the last 10 patients/rows
  print(f"Data of patient {i+51}: {last_10_patients[i]}\n")  # Displaying the last 10 patients' inflammation data per patient
````

3. For each patient, compute the average inflammation over all days. You should compute all patients' average inflammation values in single line. The result is an array.

````python
mean_patients = np.mean(arr, axis = 1)  # Computing the average inflammation across the rows, or for all patients for all days

for i in range(0, len(mean_patients)):  # Running a loop for every single patient/row
  print(f"Average inflammation value for patient {i+1}: {round(mean_patients[i], 2)}")  # Displaying the average inflammation per patient over all days
````

4. Plot all patients' averages and label the plot with "Average Inflammation of Patients".

````python
fig = plt.figure(figsize=(6, 5))  # Declaring the figure size of the plot
plt.title("Average Inflammation of Patients")  # Declaring the title of the plot
plt.plot(mean_patients)  # Ploting the average per patient
plt.xlabel("patients")  # Setting the x-axis label
plt.ylabel("average")  # Setting the y-axis label
plt.show()  # Showing the plot
````

5. For each day, compute the average inflammation over all patients. You should compute all the daily average inflammation values in a single line. The result is an array.

````python
mean_days = np.mean(arr, axis = 0)  # Computing the average inflammation across the columns, or for each day for all patients

for i in range(0, len(mean_days)):  # Running a loop for every single day/column
  print(f"Average inflammation value for day {i+1}: {round(mean_days[i], 2)}")  # Displaying the average inflammation per day over all patient
````

6. Plot all daily averages and label the plot with "Average Inflammation of Days"

````python
fig = plt.figure(figsize=(6, 5))  # Declaring the figure size of the plot
plt.title("Average Inflammation of Days")  # Declaring the title of the plot
plt.plot(mean_days)  # Ploting the average per patient
plt.xlabel("days")  # Setting the x-axis label
plt.ylabel("average")  # Setting the y-axis label
plt.show()  # Showing the plot
````


## Qestion 3 [20 marks]:
Continue with the 'inflammation-data' directory. Given the following list of file names:
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv'],
```
write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](https://i.imgur.com/XXT2Jqw.png)

````python
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']  # Storing the name of all the files we want to plot for

files = [f"/content/drive/MyDrive/Colab Notebooks/INFO 212/Week 3/Assignment 1/inflammation-data/{filename}" for filename in filenames]  # Using list comprehension to get the paths for all the above mentioned files

list_arrs = []  # Initializing an empty list to store all file data
list_arrs.append(np.loadtxt(files[0], delimiter=","))  # Loading data of first file into list
list_arrs.append(np.loadtxt(files[1], delimiter=","))  # Loading data of second file into list
list_arrs.append(np.loadtxt(files[2], delimiter=","))  # Loading data of third file into list

for arr, title in zip(list_arrs, filenames):  # For loop where "arr" runs through contents of "list_arrs" and "title" runs through filenames
    fig = plt.figure(figsize=(12, 3))  # Setting the figure size for all the plots
    plt.suptitle(title)  # Puts the title/filename for every file we are plotting for
    ax1 = fig.add_subplot(131)  # Sets the position for the first plot - Left
    ax2 = fig.add_subplot(132)  # Sets the position for the second plot - Center
    ax3 = fig.add_subplot(133)  # Sets the position for the third plot - Right

    ax1.plot(np.mean(arr, axis=0))  # Plots the average across axis 0
    ax1.set_ylabel("average")  # Setting the y-axis label
    ax2.plot(np.max(arr, axis=0))  # Plots the average across axis 0
    ax2.set_ylabel("max")  # Setting the y-axis label
    ax3.plot(np.min(arr, axis=0))  # Plots the average across axis 0
    ax3.set_ylabel("min")  # Setting the y-axis label

    plt.show()  # Showing the plot
````
