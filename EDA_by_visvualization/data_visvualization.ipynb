1: Load the LA Museum Visitors data file into a DataFrame. Create a line chart that shows how the number of visitors to each museum evolved over time. Your figure should have multiple lines (one for each museum). Discuss insights you can draw from the visualization.

```python
df1 = pd.read_csv("/Users/mustafabookwala/Desktop/Drexel/Pre-Junior/Fall 2024/INFO 212/Week 6/museum-visitors(1).csv")  # Reading .csv file to dataframe using pandas

df1['Month'] = pd.to_datetime(df1['Month'])  # Converts the Month column into a parsable Month format using the datetime function
plt.figure(figsize=(12, 6))  # Setting the figure size

for column in df1.columns[1:]:  # Runs a loop from the 2nd column till the last (For all the museums)
    plt.plot(df1['Month'], df1[column], label=column)  # Plots the line chart for each museum

plt.title("LA Museum Visitors Over Time")  # Displaying title of plot
plt.xlabel("Years")  # Displaying x-axis label of plot
plt.ylabel("Number of Visitors")  # Displaying y-axis label of plot
plt.legend()  # Displaying the legend that tells us which line is for which LA museum

plt.show()  # Displays the plot
```

1: Load the `daily-minimum-temperatures-in-me.csv` as a DataFrame. Create a line plot showing the changes of the temperatures over the time. Label the xticks with quarterly months such as `1981-01`, `1981-04`, `1981-07`, `1981-10`, `1982-01`,...`1990-01`, `1990-04`,...

```python
df2 = pd.read_csv("/Users/mustafabookwala/Desktop/Drexel/Pre-Junior/Fall 2024/INFO 212/Week 6/daily-minimum-temperatures-in-me.csv")  # Reading .csv file to dataframe

df2['Date'] = pd.to_datetime(df2['Date'], format='%m/%d/%Y')  # Converts the Date column into a parsable Month format using the datetime function
df2['Daily minimum temperatures'] = pd.to_numeric(df2['Daily minimum temperatures'], errors='coerce')  # Ensuring temperature column has only number formats

plt.figure(figsize=(12, 6))  # Setting the figure size

plt.plot(df2['Date'], df2['Daily minimum temperatures'])  # Plotting Date against Daily minimum temperatures

# Customizing x-axis ticks to show dates quarterly
xticks = []  # List for x-axis positions
xlabels = []  # List for x-axis labels
for i in range(0, len(df2), 120):  # Setting step for selecting every 120th date for quarterly ticks
    xticks.append(df2['Date'].iloc[i])  # Adding selected date positions
    xlabels.append(df2['Date'].iloc[i].strftime('%Y-%m'))  # Formatting date labels as Year-Month

plt.xticks(xticks, labels=xlabels, rotation=90)  # Setting x-axis ticks with quarterly labels and rotating 90°

# Setting y-axis ticks to show temperature range at intervals of 3°C
y_min = int(df2['Daily minimum temperatures'].min())  # Minimum temperature
y_max = int(df2['Daily minimum temperatures'].max())  # Maximum temperature
plt.yticks(range(y_min, y_max + 1, 3))  # Setting y-ticks from min to max with 5°C intervals

plt.title("Daily Minimum Temperatures Over Time")  # Displaying title of plot
plt.xlabel("Date (Quarterly)")  # Displaying x-axis label of plot
plt.ylabel("Temperature (°C)")  # Displaying y-axis label of plot

plt.show()  # Displaying the plot
```
