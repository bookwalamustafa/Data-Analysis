3.  Use seaborn.histplot() to plot histograms for each of the feature columns, colored by the species. You should have 4 histogram plots. What do you see from the plots? Explain the distributions against the species.

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']  # List of all the columns in the dataframe

for column in columns:  # Loop for each column in the above list
    plt.figure(figsize=(8, 5))  # Sets the size of the histogram
    sns.histplot(data=df, x=column, hue='species')  # Uses seaborn for the histogram. Sets the x-axis as the respective column and hue acts as the different color for each species
    plt.title(f'Histogram for {column} by species')  # Sets the title of the histogram
    plt.show()  # Displays the histogram
```

4. Scatter plot all pairs of different features and color the points with the species. You should have 6 scatter plots.

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']  # List of all the columns in the dataframe

for i in range(len(columns)):  # Loops from 0 to the end of the above list
    for j in range(i + 1, len(columns)):  # Loops from the consecutive element till the end of the list
        plt.figure(figsize=(8, 5))  # Sets the size of the scatter plot
        sns.scatterplot(data=df, x=columns[i], y=columns[j], hue='species')  # Uses seaborn for the scatterplot. Sets the x-axis and y-axis as the respective columns and hue acts as the different color for each species
        plt.title(f'Scatter Plot for {columns[i]} vs {columns[j]}')  # Sets the title of the scatterplot
        plt.show()  # Displays the scatterplot
```
